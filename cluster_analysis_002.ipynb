# Using K-means from sklearn to demonstrate clustering on the Iris dataset

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target
```

```python
# Print the feature names
print("Feature names:", iris.feature_names)

# Print the target names
print("Target names:", iris.target_names)

# Print the dimensions of the dataset
print("Dimensions of the dataset (number of samples, number of features):", iris.data.shape)

# Print the first few data points
print("First few data points:")
for i in range(5):
    print(f"Data point {i+1}: {iris.data[i]} (target: {iris.target[i]} - {iris.target_names[iris.target[i]]})")
```

```python
# Print 15 randomly selected data points
random_indices = np.random.choice(iris.data.shape[0], size=15, replace=False)
random_indices
```

```python
print("Randomly selected data points:")
for idx in random_indices:
    print(f"Data point {idx}: {iris.data[idx]} (target: {iris.target[idx]} - {iris.target_names[iris.target[idx]]})")
```

```python
# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
print("Randomly selected data points:")
for idx in random_indices:
    print(f"Data point {idx}: {X_scaled[idx]} (target: {iris.target[idx]} - {iris.target_names[iris.target[idx]]})")
```

```python
# Apply PCA to reduce the dimensionality for visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
X_pca
```

```python
print("Randomly selected data points:")
for idx in random_indices:
    print(f"Data point {idx}: {X_pca[idx]} (target: {iris.target[idx]} - {iris.target_names[iris.target[idx]]})")
```

# Elbow method to find the optimal number of clusters

```python
# Initialize an empty list to store the values of WCSS
wcss = []

# Try cluster numbers from 1 to 10
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X_scaled)
    # Append the within-cluster sum of squares to the list
    wcss.append(kmeans.inertia_)
wcss
```

```python
# Plot the elbow method graph
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='-')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.xticks(range(1, 11))
plt.grid(True)
plt.show()
```

```python
# Perform K-means clustering
elbow_n_clusters = 3
kmeans = KMeans(n_clusters=elbow_n_clusters, random_state=42) # default n-init=1
kmeans.fit(X_scaled)
y_kmeans = kmeans.predict(X_scaled)
y_kmeans
```

```python
# Plot the results
plt.figure(figsize=(10, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_kmeans, cmap='viridis', edgecolor='k', s=50)
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='x', s=200, label='Cluster Centers')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('K-means Clustering on Iris Dataset')
plt.legend()
plt.show()
```

```python
# Perform K-means clustering
elbow_n_clusters = 3
kmeans = KMeans(n_clusters=elbow_n_clusters, random_state=42, n_init=10) # we use n_init=10
kmeans.fit(X_scaled)
y_kmeans = kmeans.predict(X_scaled)
y_kmeans
```

```python
# Plot the results
plt.figure(figsize=(10, 6))

plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_kmeans, cmap='viridis', edgecolor='k', s=50)
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='x', s=200, label='Cluster Centers')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('K-means Clustering on Iris Dataset')
plt.legend()
plt.savefig('k_means_k_3_n_init_10.png')
plt.show()

```

# Test different n_init values to visualize clusters

```python
# Create a 4x4 grid of subplots
fig, axs = plt.subplots(4, 4, figsize=(16, 16))
fig.subplots_adjust(hspace=0.4, wspace=0.4)

# Iterate through different values of n_init
for i, n_init in enumerate(range(1, 17)):
    # Perform K-means clustering
    kmeans = KMeans(n_clusters=3, random_state=42, n_init=n_init)
    kmeans.fit(X_scaled)
    y_kmeans = kmeans.predict(X_scaled)
    
    # Plot the clustering results
    ax = axs[i // 4, i % 4]
    ax.scatter(X_pca[:, 0], X_pca[:, 1], c=y_kmeans, cmap='viridis', edgecolor='k', s=50)
    centers = kmeans.cluster_centers_
    ax.scatter(centers[:, 0], centers[:, 1], c='red', marker='x', s=200, label='Cluster Centers')
    ax.set_title(f'n_init = {n_init}')
    ax.set_xlabel('Principal Component 1')
    ax.set_ylabel('Principal Component 2')
    ax.legend()

# Show the plot
plt.tight_layout()
plt.show()
```
